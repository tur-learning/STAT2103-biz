# üé≤ **5.1 The Concept of a Sampling Distribution** üìä

### **Why Do We Need Sampling Distributions?** ü§î
In **Chapter 4**, we assumed that we already **knew the probability distribution** of a random variable. That allowed us to compute **expected values, variances, and probabilities**.  
### üöÄ **But in real-world scenarios, we often don‚Äôt know these distributions!**
Instead, we use **samples** to estimate unknown **population parameters** (like the mean, variance, or proportion).  

For example:
- **Unionization poll** üìã: We assumed 60% of employees favor unionization, but the real percentage is unknown.
- **Car gas mileage** üöó‚õΩ: We used **mean = 27 mpg, standard deviation = 3 mpg**, but the true values might be different.

#### That‚Äôs why we introduce **sampling distributions**! üî•  

---

### **üéØ What is a Sampling Distribution?**
A **sampling distribution** is the probability distribution of a **sample statistic** (such as the sample mean, median, or variance) **computed from repeated random samples** of the same size.  

####üí° **Key Idea**:
If we **repeatedly draw random samples** and calculate a statistic (like the sample mean), these values will **themselves form a distribution**! üé≤üìä

####üëâ **Example:**
Suppose a factory produces electronic components, and the **length of a connector module** is normally distributed with:
- Population **mean**: $ \mu = 5.0 $ cm
- Population **standard deviation**: $ \sigma = 0.005 $ cm

If we **randomly sample** 25 modules and compute their **sample mean $ \bar{x} $**, the value of $ \bar{x} $ will **change** across different samples! If we **repeat this sampling experiment many times**, we will get a **distribution of sample means**, which is the **sampling distribution of $ \bar{x} $**. üéØ


---

## **üß™ Example 1: Sampling Distribution of the Sample Mean** üé≤

Let's **simulate** a sampling distribution of the **sample mean** when rolling a **fair die**. üé≤

- A **single die roll** produces numbers **1 to 6** (each equally likely).
- The **true population mean** is:  
  \begin{align}
  \mu = \frac{1+2+3+4+5+6}{6} = 3.5
  \end{align}
- We **roll a die 3 times**, compute the **sample mean** $ \bar{x} $, and repeat this experiment **10,000 times** to see the **distribution of $ \bar{x} $**.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters üé≤
num_samples = 10000  # Number of times we take a sample
sample_size = 3  # Number of rolls per sample

# Generate the sampling distribution of the sample mean üéØ
sample_means = [np.mean(np.random.randint(1, 7, sample_size)) for _ in range(num_samples)]

# Plot the histogram üìä
plt.hist(sample_means, bins=20, density=True, alpha=0.6, color='blue', edgecolor="black")
plt.axvline(np.mean(sample_means), color='red', linestyle='dashed', linewidth=2, label=f"Mean = {np.mean(sample_means):.2f}")
plt.xlabel("Sample Mean (ùë•ÃÑ)")
plt.ylabel("Frequency")
plt.title("Sampling Distribution of the Sample Mean (Die Rolls)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

### **üîç Interpretation:**
- The **sampling distribution of $ \bar{x} $** is approximately **normal**, even though the original die rolls were **uniformly distributed**! üé≤‚û°Ô∏èüìä
- The **mean of the sample means** is **very close to 3.5** (the true population mean). ‚úÖ
- **Key Insight:** The sample mean $ \bar{x} $ tends to **cluster around $ \mu $** as the number of samples increases.  

---

## **üß™ Example 2: Sampling Distribution of Sample Median** üî¢

Now let‚Äôs compare **sample mean $ \bar{x} $ and sample median $ m $** to see which is a better estimator of the population mean. üî•  

We'll simulate **steel sheet thickness** data from a **uniform distribution** between **150 and 200 mm** and compare the sampling distributions of:
- **Sample mean $ \bar{x} $**
- **Sample median $ m $**

```python
# Parameters
num_samples = 10000
sample_size = 11

# Generate sampling distributions üìä
sample_means = [np.mean(np.random.uniform(150, 200, sample_size)) for _ in range(num_samples)]
sample_medians = [np.median(np.random.uniform(150, 200, sample_size)) for _ in range(num_samples)]

# Plot histograms üìä
plt.figure(figsize=(12,5))

# Sample Mean Distribution
plt.subplot(1,2,1)
plt.hist(sample_means, bins=20, density=True, alpha=0.6, color='blue', edgecolor="black")
plt.axvline(np.mean(sample_means), color='red', linestyle='dashed', linewidth=2, label=f"Mean = {np.mean(sample_means):.2f}")
plt.xlabel("Sample Mean " + r"($\bar{x}$)")
plt.ylabel("Frequency")
plt.title("Sampling Distribution of the Sample Mean")
plt.legend()
plt.grid(alpha=0.3)

# Sample Median Distribution
plt.subplot(1,2,2)
plt.hist(sample_medians, bins=20, density=True, alpha=0.6, color='green', edgecolor="black")
plt.axvline(np.mean(sample_medians), color='red', linestyle='dashed', linewidth=2, label=f"Mean = {np.mean(sample_medians):.2f}")
plt.xlabel("Sample Median (m)")
plt.ylabel("Frequency")
plt.title("Sampling Distribution of the Sample Median")
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

### **üîç Interpretation:**
- The **sample mean $ \bar{x} $** clusters **closer to the population mean** than the sample median $ m $. ‚úÖ
- The **sample median has a larger spread** (more variability).  
- **Conclusion:** The **sample mean** is a **better estimator** of the population mean $ \mu $ than the **sample median** in this case. üî•

---

## **üéØ Key Takeaways:**
| üé≤ **Concept** | üìå **Explanation** |
|--------------|----------------|
| **Sampling Distribution** | The distribution of a **sample statistic** (e.g., sample mean) across repeated samples. |
| **Law of Large Numbers** | As sample size **increases**, the sample mean $ \bar{x} $ **approaches** the population mean $ \mu $. |
| **Central Limit Theorem** | Even if the **original population is not normal**, the **sampling distribution of the sample mean** will be **approximately normal** if $ n $ is large. ‚úÖ |
| **Choosing an Estimator** | The **sample mean** is typically a **better** estimator of $ \mu $ than the **sample median** because it has **less variability**. |

# **üìä 5.2 Properties of Sampling Distributions: Unbiasedness and Minimum Variance üéØ**  

In this section, we'll explore two key properties of **point estimators**‚Äî**unbiasedness** and **minimum variance**‚Äîwhich help us determine the **best** statistic to use when estimating a population parameter. üöÄ  

---

### **üéØ What is a Point Estimator?**
A **point estimator** is a formula or rule that calculates a **single number** to estimate an **unknown population parameter**.  

Examples:

| **Parameter** (Population) | **Estimator** (Sample) |
|--------------------------|-------------------------|
| **Mean** $ \mu $ | Sample mean $ \bar{x} $ |
| **Variance** $ \sigma^2 $ | Sample variance $ s^2 $ |
| **Standard deviation** $ \sigma $ | Sample standard deviation $ s $ |
| **Proportion** $ p $ | Sample proportion $ \hat{p} $ |


### **üìå But... Are All Estimators Good?**
NO! ‚ö†Ô∏è Some estimators tend to **overestimate** or **underestimate** a parameter. Others have **high variability**, making them unreliable.  
This is why we look at **two key properties**:  

1Ô∏è‚É£ **Unbiasedness** (Does the estimator give correct values on average?)  
2Ô∏è‚É£ **Minimum variance** (Does the estimator have the least spread in repeated sampling?)  

---

## **üéØ 1. Unbiased Estimators (No Systematic Error!)**
A statistic is **unbiased** if its expected value (mean of its sampling distribution) is equal to the **true population parameter**.

üí° **Mathematically:**  

\begin{align}
E[\hat{\theta}] = \theta
\end{align}

where $ \hat{\theta} $ is an estimator of the population parameter $ \theta $.  

üîπ **Example: Unbiased vs. Biased Estimator**  
Imagine we have **two estimators, A and B, for $ \mu $**:  
- **Estimator A** is centered around $ \mu $ (unbiased).  
- **Estimator B** consistently overestimates $ \mu $ (biased).  

üìä **Visualization:**  


```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated sampling distributions
np.random.seed(42)
mu = 50  # True population mean

# Unbiased estimator: values centered around 50
unbiased_estimator = np.random.normal(mu, 5, 1000)

# Biased estimator: values shifted to the right (overestimation)
biased_estimator = np.random.normal(mu + 5, 5, 1000)

# üìä Plot the distributions
plt.hist(unbiased_estimator, bins=30, alpha=0.6, label="Unbiased Estimator", color='blue', edgecolor="black")
plt.hist(biased_estimator, bins=30, alpha=0.6, label="Biased Estimator", color='red', edgecolor="black")
plt.axvline(mu, color='black', linestyle='dashed', linewidth=2, label="True Mean (Œº)")
plt.xlabel("Estimator Value")
plt.ylabel("Frequency")
plt.title("Unbiased vs. Biased Estimator")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

**üîç Interpretation:**  
- The **blue distribution** (unbiased) is centered at $ \mu $, meaning on average, it gives correct estimates! ‚úÖ  
- The **red distribution** (biased) is shifted to the right, meaning it **systematically overestimates** $ \mu $. üö®  

**Key Takeaway:**  
###üí° **Always prefer an unbiased estimator when possible!**  

---

## **üéØ 2. Minimum Variance (Least Spread Wins!)**
Among **unbiased** estimators, the best one is the one with the **smallest variance** (less variability in repeated samples).  

üìä **Visualization: Two Unbiased Estimators, One with Less Variability**  


```python
# Two unbiased estimators with different variances
low_variance = np.random.normal(mu, 3, 1000)  # Smaller spread
high_variance = np.random.normal(mu, 8, 1000)  # Larger spread

# Plot the distributions
plt.hist(low_variance, bins=30, alpha=0.6, label="Low Variance", color='blue', edgecolor="black")
plt.hist(high_variance, bins=30, alpha=0.6, label="High Variance", color='orange', edgecolor="black")
plt.axvline(mu, color='black', linestyle='dashed', linewidth=2, label="True Mean (Œº)")
plt.xlabel("Estimator Value")
plt.ylabel("Frequency")
plt.title("üìä Unbiased Estimators with Different Variances")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

**üîç Interpretation:**  
- Both **blue and orange distributions** are centered at $ \mu $ (both are unbiased).  
- The **blue estimator has less spread** (smaller standard deviation), meaning it provides **more consistent estimates**. ‚úÖ  
- The **orange estimator has high variability**, meaning it might give values **far from $ \mu $ more often**. üö®  

**Key Takeaway:**  
### üí° Among **unbiased** estimators, choose the one with the **smallest variance**!  

This estimator is called the **Minimum-Variance Unbiased Estimator (MVUE)**. üéØ  

---

## **üß™ Example: Biased and Unbiased Estimators in Craps üé≤**  

In **craps**, a **\$5 bet** on the come-out roll can result in three possible outcomes:  
- Win **\$5** - probability **0.25**
- Lose **\$5** - probability **0.10**
- No change (**\$0**) - probability **0.65**

Let‚Äôs check if the **sample mean $ \bar{x} $ and median $ m $** are unbiased.  

```python
# Possible sample statistics and their probabilities
x_values = np.array([0, 1.67, 3.33, 5])
p_x = np.array([0.4225, 0.3375, 0.1875, 0.0525])  # Given probabilities

# Compute expected values
E_x_bar = np.sum(x_values * p_x)  # Expected value of sample mean
E_median = (0 * 0.42) + (5 * 0.58)  # Expected value of sample median

print(f"üìä Expected Value of Sample Mean: {E_x_bar:.2f}")
print(f"üìä Expected Value of Sample Median: {E_median:.2f}")
```

**üîç Interpretation:**  
- The **sample mean $ \bar{x} $ is unbiased** because its expected value equals the population mean. ‚úÖ  
- The **sample median $ m $ is biased** because its expected value **differs** from the population mean. ‚ùå  

### üîπ **Conclusion:** The sample mean $ \bar{x} $ is a **better estimator** of the population mean. üéØ  


---

## **üìå Summary Table: Choosing the Best Estimator**
| **Property** | **Good Estimator** ‚úÖ | **Bad Estimator** ‚ùå |
|-------------|----------------------|----------------------|
| **Unbiasedness** üéØ | Centered at the true parameter ($E[\hat{\theta}] = \theta$) | Systematically over/underestimates parameter |
| **Low Variability** üìâ | Small standard deviation (consistent estimates) | Large standard deviation (estimates vary a lot) |
| **MVUE (Best Choice!)** üèÜ | Unbiased **+** Minimum Variance | Biased or High Variance |

---

## **üî• Final Takeaways**
‚úÖ A **good estimator** should be **unbiased** and have the **smallest variance**.  
‚úÖ If you have two unbiased estimators, choose the one with the **smaller standard deviation**.  
‚úÖ The **sample mean $ \bar{x} $** is usually a **better estimator** than the **sample median $ m $**.  
‚úÖ **Minimum-Variance Unbiased Estimator (MVUE)** is the **gold standard**! üèÜ

## **üìä 5.3 The Sampling Distribution of the Sample Mean and the Central Limit Theorem üéØ**  


We‚Äôve talked about **sampling distributions** before, but now we‚Äôre introducing a game-changer: the **Central Limit Theorem (CLT)!** üöÄ  
This theorem **guarantees** that, for large enough sample sizes, the sampling distribution of the sample mean **follows a normal distribution**‚Äîeven when the original population is NOT normal! ü§Ø  

---

## **üéØ What is the Sampling Distribution of $ \bar{x} $?**
When we take a **random sample** of size $ n $ from a population with:  
‚úÖ Mean $ \mu $  
‚úÖ Standard deviation $ \sigma $  

The **sampling distribution of the sample mean** $ \bar{x} $ has:  
1Ô∏è‚É£ **Mean**: $ E[\bar{x}] = \mu $ (same as the population mean)  
2Ô∏è‚É£ **Standard Deviation** (Standard Error of the Mean):  
\begin{align}
\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}}
\end{align}
3Ô∏è‚É£ **Shape**:  
- If the population is normal, **$ \bar{x} $ is also normal** ‚úÖ  
- If the population is not normal, **$ \bar{x} $ is still approximately normal for large $ n $!** üéâ  

---

## **üî• The Central Limit Theorem (CLT)**
**Theorem:**  
If we take **a random sample of size $ n $** from *any* population (even non-normal ones!), then for **large $ n $, the sampling distribution of $ \bar{x} $ is approximately normal.**  


### üí° **Key takeaway**:  
Even if the population **is skewed or weird**, the sample mean $ \bar{x} $ will be **normally distributed for large $ n $**! üéØ   

### **How Large is "Large Enough"?**
- If the population is **already normal** ‚Üí CLT works **for any $ n $** üìà  
- If the population is **skewed** ‚Üí $ n \geq 30 $ is usually enough üìä  
- If the population is **heavily skewed** ‚Üí Need an even larger $ n $! ‚è≥

---

## **üìä Example 1: Visualizing the CLT with Skewed Populations**  

Let‚Äôs simulate this and see how the CLT works! We‚Äôll:
1. Generate a **skewed** population.  
2. Take random samples of **different sizes**.  
3. Show how the **sampling distribution of $ \bar{x} $ becomes normal** as $ n $ increases.  

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate a right-skewed population (Exponential Distribution)
np.random.seed(42)
population = np.random.exponential(scale=2, size=10000)  # Skewed population

sample_sizes = [5, 10, 30, 50, 100]  # Different sample sizes
fig, axes = plt.subplots(1, len(sample_sizes), figsize=(20, 4))

# Simulate sampling distributions for different n
for i, n in enumerate(sample_sizes):
    sample_means = [np.mean(np.random.choice(population, size=n)) for _ in range(1000)]

    # üìä Plot histogram
    axes[i].hist(sample_means, bins=30, density=True, alpha=0.6, color="blue", edgecolor="black")

    # Overlay a normal curve
    mu, sigma = np.mean(sample_means), np.std(sample_means)
    x = np.linspace(min(sample_means), max(sample_means), 100)
    axes[i].plot(x, stats.norm.pdf(x, mu, sigma), color="red", linewidth=2)

    axes[i].set_title(f"n = {n}")

plt.suptitle("Central Limit Theorem in Action (Skewed Population)")
plt.show()
```

**üîç Interpretation:**  
- For **$ n=5 $** ‚Üí The distribution is still **skewed**! üö®  
- For **$ n=10 $** ‚Üí It‚Äôs getting closer to normal... üßê  
- For **$ n=30 $** ‚Üí Looks **almost normal** ‚úÖ  
- For **$ n=100 $** ‚Üí The sampling distribution is **fully normal**! üéØ  

**Conclusion:**  
###üí° For **skewed populations**, the sample mean $ \bar{x} $ still **follows a normal distribution for large $ n $**!  

---

## **üéØ Example 2: Applying the CLT to a Battery Life Problem**
A battery manufacturer claims that its batteries last **$ \mu = 54 $ months** on average, with a **standard deviation of $ \sigma = 6 $ months**.  

A **consumer group** tests **$ n = 50 $** batteries and finds the average lifespan.  
We want to find the probability that the **sample mean** battery life is **less than 52 months**.  

#### üí° **Solution:**  
1Ô∏è‚É£ **Find the mean and standard deviation of $ \bar{x} $**:  
   \begin{align}
   \mu_{\bar{x}} = \mu = 54, \quad \sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}} = \frac{6}{\sqrt{50}} = 0.849
   \end{align}
   
2Ô∏è‚É£ **Compute the z-score**:  
   \begin{align}
   z = \frac{52 - 54}{0.849} = \frac{-2}{0.849} = -2.36
   \end{align}

3Ô∏è‚É£ **Find the probability using the standard normal table**:  
   \begin{align}
   P(Z < -2.36) \approx 0.0091
   \end{align}

#### üìå **Interpretation:**  
- The probability of getting a sample mean of **less than 52 months** is **0.91%**.  
- Since this is **very small**, if the sample mean **actually turns out to be 52**, we might suspect the manufacturer **is lying!** üò±  


---

## **üìå Summary Table: Key CLT Takeaways**
| **Concept** | **Explanation** |
|------------|---------------|
| **Sampling Distribution of $ \bar{x} $** | Follows a normal distribution for **large** $ n $, even if the population is not normal. |
| **Mean of $ \bar{x} $** | $ E[\bar{x}] = \mu $ (same as the population mean). ‚úÖ |
| **Standard Deviation of $ \bar{x} $** | $ \sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}} $ (smaller for larger $ n $). üìâ |
| **When is $ \bar{x} $ Normally Distributed?** | ‚úÖ **Always if the population is normal.**<br>‚úÖ **Approximately normal if $ n \geq 30 $.** |
| **Why is the CLT Important?** | Lets us use **normal probability methods** even when populations are not normal! üéØ |


---

## **üî• Final Takeaways**
‚úÖ The **Central Limit Theorem (CLT)** allows us to assume **normality** for $ \bar{x} $ even when the population is **not normal**! üìä  
‚úÖ The **larger the sample size $ n $, the better the normal approximation**. üöÄ  
‚úÖ The **sampling distribution of $ \bar{x} $** has a smaller standard deviation than the original population. üìâ  
‚úÖ If an observed sample mean is **very different** from the expected value, it might indicate **the population claim is false**! ‚ùå

## **üìä 5.4 The Sampling Distribution of the Sample Proportion üéØ**  

So far, we've discussed the **sampling distribution of the sample mean** $ \bar{x} $, but now let's dive into another super important statistic:
#### the **sample proportion** $ \hat{p} $! üöÄ  


This comes in handy when dealing with **categorical** data (e.g., Yes/No, Pass/Fail, Approve/Disapprove).
#### When estimating the **population proportion** $ p $, the sample proportion $ \hat{p} $ is our best friend! ü§ù

---

## **üéØ What is the Sampling Distribution of $ \hat{p} $?**  
When we take a **random sample** of size $ n $ from a population with:
‚úÖ **True proportion of "successes"**: $ p $  
‚úÖ **True proportion of "failures"**: $ 1 - p $  

The **sampling distribution of $ \hat{p} $** has:  
1Ô∏è‚É£ **Mean**: $ E[\hat{p}] = p $ (same as the population proportion)  
2Ô∏è‚É£ **Standard Deviation** (Standard Error of $ \hat{p} $):  
\begin{align}
\sigma_{\hat{p}} = \sqrt{\frac{p(1 - p)}{n}}
\end{align}
3Ô∏è‚É£ **Shape**:  
- If $ n $ is **large**, the distribution of $ \hat{p} $ is **approximately normal**.  
- **Rule of thumb** for normality:  
  ‚úÖ $ np \geq 5 $  
  ‚úÖ $ n(1 - p) \geq 5 $  

#### If both conditions hold, we can **safely use the normal approximation**! üìà  


---

## **üî• Example 1: Simulating the Sampling Distribution of $ \hat{p} $**  

Let‚Äôs **simulate** the process of taking multiple samples and computing $ \hat{p} $ each time! We'll:  
1. Assume **60% of workers favor unionization** $ (p = 0.6) $  
2. Take **samples of 100 workers**  
3. Compute $ \hat{p} $ for **each sample**  
4. Plot the **sampling distribution**  

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define parameters
p = 0.6  # True population proportion
n = 100  # Sample size
num_samples = 1000  # Number of samples

# Generate samples from a binomial distribution
samples = np.random.binomial(n, p, size=num_samples) / n  # Compute sample proportions

# üìä Plot histogram
plt.figure(figsize=(10, 5))
plt.hist(samples, bins=30, density=True, alpha=0.6, color="blue", edgecolor="black")

# Overlay normal approximation
mu, sigma = np.mean(samples), np.std(samples)
x = np.linspace(min(samples), max(samples), 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), color="red", linewidth=2, label="Normal Approximation")

plt.axvline(p, color="black", linestyle="dashed", label=f"True Proportion (p = {p})")
plt.title("Sampling Distribution of Sample Proportion $ \hat{p} $")
plt.xlabel("Sample Proportion")
plt.ylabel("Density")
plt.legend()
plt.show()
```


**üîç Interpretation:**  
- The **sampling distribution** of $ \hat{p} $ is approximately **normal**! ‚úÖ  
- The **mean** of $ \hat{p} $ is very close to **$ p = 0.6 $** üéØ  
- The **spread of $ \hat{p} $** depends on $ n $ (larger $ n $ means less variability!) üìâ  

#### **Conclusion:**  
üí° For **large $ n $**, the **sample proportion** follows a **normal distribution** even though the population data is **categorical!** ü§Ø  

---

## **üéØ Example 2: Finding a Probability Involving $ \hat{p} $**  

Imagine we're testing **100 workers** in a company where **60% support unionization**.  
**Question:** ‚ùì  
What is the probability that **less than 50% of the sample** supports unionization?  

#### üí° **Solution:**  
1Ô∏è‚É£ **Check if normal approximation applies:**  
   \begin{align}
   np = 100(0.6) = 60 \quad \text{(‚úÖ OK)}
   \end{align}
   \begin{align}
   n(1 - p) = 100(0.4) = 40 \quad \text{(‚úÖ OK)}
   \end{align}
   Since both conditions are met, we can **use the normal approximation**! üìà  

2Ô∏è‚É£ **Find the mean and standard deviation of $ \hat{p} $:**  
   \begin{align}
   \mu_{\hat{p}} = p = 0.6
   \end{align}
   \begin{align}
   \sigma_{\hat{p}} = \sqrt{\frac{p(1 - p)}{n}} = \sqrt{\frac{0.6(0.4)}{100}} = 0.049
   \end{align}

3Ô∏è‚É£ **Compute the z-score for $ \hat{p} = 0.5 $:**  
   \begin{align}
   z = \frac{0.5 - 0.6}{0.049} = \frac{-0.1}{0.049} = -2.04
   \end{align}

4Ô∏è‚É£ **Find the probability using the normal table:**
