# 1.2 Types of Statistical Applications in Business

Statistics in business involves two key processes:

- **Descriptive Statistics**: Summarizes and organizes data using
  1. numerical methods
  2. graphical methods
  
  to reveal **patterns** and present the information effectively.
- **Inferential Statistics**: Uses sample data to make
  1. generalizations
  2. predictions
  3. decisions
  
  about a larger set of data.

## Key Definitions

**Sampling**: The process of selecting a subset of data to represent the whole population.

*For example, analyzing the ages of a sample of customers to estimate the average age of all customers.*

Descriptive Statistics: Exploring and presenting data patterns using methods like charts, graphs, and numerical summaries.

Inferential Statistics: Drawing conclusions or making predictions about a population based on sample data.

We will discuss **both** descriptive and inferential statistics in the following chapters, but the focus here is on inference.

Since 1917, the Girl Scouts of America have been selling boxes of cookies. In 2017, there were 12 varieties for sale:

Each of the approximately 150 million boxes of Girl Scout cookies sold in 2017 was classified by variety. Results are summarized here below.

**References**:
- [Girls Scout](http://www.girlscouts.org/)
- [Cookie Flavors](https://www.girlscouts.org/en/cookies/cookie-flavors.html)

```python
# Data
import matplotlib.pyplot as plt


# Data
flavors = [
    "ThinMints", "Trefoils", "Do-si-dos", "other",
    "Samoas", "Tagalongs"
]
percentages = [25, 9, 11, 23, 19, 13]

# Plot
fig, ax = plt.subplots(figsize=(5, 5))  # Increase figure size
ax.pie(
    percentages,
    labels=flavors,
    autopct='%1.1f%%',
    startangle=140,
    labeldistance=1.1  # Move labels slightly further out
)
plt.title("Best-selling Girl Scout cookies")
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

# 1.3 Fundamental Elements of Statistics

Statistics is invaluable for studying populations and making informed decisions based on data. Understanding the key elements of statistics helps in effectively analyzing and interpreting information.

## Key Concepts

1. **Experimental** (or **Observational**) **Unit**: The object (e.g., person, thing, transaction) on which data is collected.
2. **Population**: The complete set of units of interest. e.g. all employees in a company, all cars produced in a year, all sales transactions at a store.

Notice that these population examples are sets (groups) of people, sets of objects, set of transactions, and set of events.

---

In studying a population, we focus on one or more characteristics or properties of the experimental units in the population.

- **Variable**: A characteristic or property of an experimental unit (e.g., age, income, error status).

The name variable is derived from the fact that any particular characteristic may vary among the experimental units in a population.

In studying a particular variable, it is helpful to be able to obtain a numerical representation for it. Often, however, numerical representations are not readily available.

1. **Measurement**: The process of assigning numbers to variables (e.g., age in years, error status as 1 or 0).
2. **Census**: Measuring a variable for every unit in the population.

A reasonable alternative would be to select and study a subset (or portion) of the units in the population.
- **Sample**: A subset of the population used for analysis.

For example, a company is being audited for invoice errors.

Instead of examining all 15,472 invoices produced by the company during a given year, an auditor may select and examine a sample of just 100 invoices. If he is interested in the variable “invoice error status,” he would record (measure) the status (error or no error) of each sampled invoice.

image.png

After the variable(s) of interest for every experimental unit in the sample (or population) is (are) measured, the data are analyzed, either by descriptive or by inferential statistical methods.

The auditor, for example, may be interested only in describing the error rate in the sample of 100 invoices.

More likely, however, he will want to use the information in the sample to make inferences about the population of all 15,472 invoices.

- **Statistical Inference**: An estimate, prediction, or generalization about a population based on a sample.

```python
# Exercise 2: Estimating population metrics

# Sample data
sample_size = 100
sample_errors = 12

# Population size
population_size = 15472

# Calculate the sample error rate
sample_error_rate = sample_errors / sample_size
print(f"Sample Error Rate: {sample_error_rate:.2%}")

# Estimate total errors in the population
estimated_population_errors = sample_error_rate * population_size
print(f"Estimated Total Errors in Population: {int(estimated_population_errors)}")
```

That is, we use the information contained in the sample to learn about the larger population. The auditor’s inference about the quality of the firm’s invoices can be used in deciding whether to modify the firm’s billing operations.

*The terms population and sample are often used to refer to the sets of measurements themselves, as well as to the units on which the measurements are made. When a single variable of interest is being measured, this usage causes little confusion. But when the terminology is ambiguous, we’ll refer to the measurements as population data sets and sample data sets, respectively.*
